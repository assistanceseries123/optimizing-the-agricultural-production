```python
#Lets try to cluster these crops
#Lets select all the columns from the dataset
x=df.loc[:,['K','N','P','rainfall','humidity','temperature','ph']]
print(x.shape)
x_data=pd.DataFrame(x)
x_data.head()
```

```python
#Lets determinre the optimum number of the cluster with in the dataset
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    km.fit(x)
    wcss.append(km.inertia_)
#Lets plot the above result
plt.rcParams['figure.figsize']=(17,6)
plt.style.use('fivethirtyeight')
plt.plot(range(1,11),wcss)
plt.xlabel("No.of cluster")
plt.ylabel("wcss")
plt.title("The Elbow method",fontsize=10)
plt.show()
```

```python
#Lets implement the kmeans algorithm to perform the clustering analysis
km=KMeans(n_clusters=4,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_means=km.fit_predict(x)
#Lets find the result
a=df['label']
y_means=pd.DataFrame(y_means)
z=pd.concat([y_means,a],axis=1)
z=z.rename(columns={0:'cluster'})
#Lets check the cluster of each crops
print("Lets check the results after applying the cluster\n")
print("Crops in the first cluster:",z[z['cluster']==0]['label'].unique())
print("Crops in the second cluster:",z[z['cluster']==1]['label'].unique())
print("Crops in the third cluster:",z[z['cluster']==2]['label'].unique())
print("Crops in the fourth cluster:",z[z['cluster']==3]['label'].unique())
```

# Predictive modelling

```python
y=df['label']
x=df.drop(['label'],axis=1)
```

```python
print(x.shape)
```

```python
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
models={
    "Logistice Regression":LogisticRegression(),
    "Decision tree":DecisionTreeClassifier(),
    "RandomForest":RandomForestClassifier()
}
```

```python
model1=LogisticRegression(multi_class='auto')
model1.fit(X_train,y_train)
y_pred=model1.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,cmap='viridis')
```
